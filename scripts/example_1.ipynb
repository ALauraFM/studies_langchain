```python
from langchain_openai import OpenAI, ChatOpenAI
import os
import yaml
```

```python
with open("config.yaml", "r") as config_file:
    config = yaml.safe_load(config_file)
    os.environ["OPENAI_API_KEY"] = config["OPENAI_API_KEY"]
```

Completion

```python
openai = OpenAI(model_name='gpt-3.5-turbo-instruct')
temperature = 1
freqeuency_penalty = 1
presence_penalty = 1
max_tokens = 500
n = 1
seed = 123
response = openai.invoke(input='Write a poem about AI in the style of Shakespeare.', temperature=temperature, 
                         frequency_penalty=freqeuency_penalty, 
                         presence_penalty=presence_penalty, max_tokens=max_tokens, 
                         n=n, seed=seed)


print(response)
```

Chat

```python
openai = ChatOpenAI(model_name='gpt-3.5-turbo')
messages = [
      {"role": "system", "content": "You are an assistant that helps with cinema facts."},
      {"role": "user", "content": "Which movie has won more oscars on the cinema history?"}
      ]
response = openai.invoke(messages)
#full response
print(response)
#only the content of the response
print(response.content)
```
