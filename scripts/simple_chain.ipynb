```python
from langchain_openai import OpenAI
from langchain_core.prompts import PromptTemplate
import yaml
import os
```

```python
with open("config.yaml", "r") as config_file:
    config = yaml.safe_load(config_file)
    os.environ["OPENAI_API_KEY"] = config["OPENAI_API_KEY"]
```

```python
openai = OpenAI(model_name='gpt-3.5-turbo-instruct', temperature=0)
```

Simple

```python
prompt_template = PromptTemplate.from_template("Describe in a few words who was {person} and his most famous role.")
runnable_sequence = prompt_template | openai
output = runnable_sequence.invoke({"person": "Ian McKellen"})
```

```python
print("Output:\n", output)
```
